```python
def myThresh(img,thresh):
    row,cols=img.shape

    for i in range(row):
        for j  in range(cols):
            if(img[i,j]>thresh):
                img[i,j]=255.0
            else:
                img[i,j]=0.0
    return img
```

```python
import cv2
import numpy as np

# in OpenCV, h value: 0-180 (Red 0-60,Green 60-120, Blue 120-180)
# s value: 0-255
# v value: 0-255

#cap = cv2.VideoCapture(1)

def nothing(x):
    pass
# Creating a window for later use
cv2.namedWindow('result')

# Starting with 100's to prevent error while masking
h,s,v = 100,100,100

# Creating track bar
cv2.createTrackbar('h', 'result',0,179,nothing)
cv2.createTrackbar('s', 'result',0,255,nothing)
cv2.createTrackbar('v', 'result',0,255,nothing)

while(1):

    frame = cv2.imread('colors3.jpg')
    #ret,frame=cap.read()
    frame=cv2.resize(frame,(650,350))

    #converting to HSV
    hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)

    # get info from track bar and appy to result
    h = cv2.getTrackbarPos('h','result')
    s = cv2.getTrackbarPos('s','result')
    v = cv2.getTrackbarPos('v','result')

    # Normal masking algorithm
    lower_color = np.array((h,s,v))
    upper_color = np.array((85,255,255)) #define the upper color range (Found from the internet)

    # green - lower 29,54,129
    
    mask = cv2.inRange(hsv,lower_color, upper_color)

    result = cv2.bitwise_and(frame,frame,mask = mask)

    cv2.imshow('result',result)

    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break

cv2.destroyAllWindows()
```

#### Detect the plaint height

```python
import cv2
import numpy as np

frame=cv2.imread('plants/D14.png')
hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)

upper_color=np.array((50,255,255))   # H,V set maxium value
lower_color=np.array((23,37,110))    # Using following UI 

mask=cv2.inRange(hsv,lower_color,upper_color)  # Only blue color object is highlited

contours,hierarchy=cv2.findContours(mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

for i,cnt in enumerate(contours):

    area=cv2.contourArea(cnt)

    if(area>500):
        # cv2.drawContours(frame,[cnt],-1,(0,0,255),3) 
        x,y,w,h=cv2.boundingRect(cnt)  # this unit is pixel
        # draw the rect 
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),3)

        # TO get the height of plaint
        cv2.putText(frame,'height:'+str(h),(x+w+10,y+h),cv2.FONT_HERSHEY_SIMPLEX,0.75,(255,255,255),2)




cv2.imshow('frame',frame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Example 2 - Lane Detection using Edge Detection using Sobel Kernels

```python
import cv2
import numpy as np

source = cv2.VideoCapture("road_drive.avi")

kernel = np.array([[-1,0,1],
                   [-2,0,2],
                   [-1,0,1]], dtype=np.float32)

while True:
    ret, img = source.read()

    # mark middle line in img
    h,w=img.shape[:2]
    frameCx=w//2
    
    

    if not ret:
        break   

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    conv = cv2.filter2D(gray, -1, kernel)
    ret, thresh = cv2.threshold(conv, 100, 255, 0) # to set black and white only no gray colors

    
    thresh[:120,:]=0 # set balck
    
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    
    for cnt in contours:
        area=cv2.contourArea(cnt)
        #print(area)
        if(area>50):
            #cv2.drawContours(img, [cnt], -1, (0,255,255), 3)  # detect other things 
            
            # x,y,w,h = cv2.boundingRect(cnt)
            # img = cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)  
            
            rect = cv2.minAreaRect(cnt)
            box = cv2.boxPoints(rect)
            box = box.astype(int)
            cv2.drawContours(img,[box],0,(0,0,255),2)

            #draw the line
            cv2.line(img,(frameCx,0),(frameCx,h),(0,255,0),2)

            ## To get center of contour
            M=cv2.moments(cnt)
            cx = int(M["m10"] / M["m00"])
            cy = int(M["m01"] / M["m00"])

            cv2.line(img,(cx,cy),(frameCx,cy),(0,255,0),2)

            
            

            

    cv2.imshow("LIVE", img)
    cv2.imshow("GRAY", gray)
    cv2.imshow("EDGE", conv)
    cv2.imshow("thresh",thresh)

    


    

    if cv2.waitKey(1) & 0xFF == 27:  # ESC key
        break

source.release()
cv2.destroyAllWindows()
```
