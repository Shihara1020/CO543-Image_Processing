## Task 01 - View the Image as an Array of Numbers

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img=cv2.imread("cats.jpg")
img_rgb= img[:,:,::-1]

plt.imshow(img_rgb)
plt.title("Original Image")
plt.show()

print("5x5 Pixel Patch (BGR values):")
print(img[100:105, 100:105])
```

#### Short Answers :

##### 1. Pixel Values :
Each pixel is represented as triplet [B,G,R] (Blue,Green,Red), where values each value ranges from 0-255.These values indicate the intensity of each color channel at that pixel.

##### 2. Relation to Image :
Higher pixel values produce brighter colors, while lower values result in darker regions. When pixel values change gradually across the image, they create smooth gradients, which appear as natural shading. Repeating numerical patterns across pixels form visible structures such as edges, objects, and textures when the image is displayed.

##### 3. Why Computers Use Numbers :
Computers cannot interpret images visually like humans. Instead, they work with numerical data and perform mathematical operations on pixel values. Representing images as arrays of numbers allows computers to process images for tasks such as enhancement, filtering, object detection, and recognition.

## Task 02 - Convert to Grayscale and Inspect Structure

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img=cv2.imread("cats.jpg")
img_rgb= img[:,:,::-1]
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


plt.imshow(gray,cmap='gray')
plt.title("Grayscale Image")
plt.show()

print("5x5 Grayscale Pixel Patch (BGR values):")
print(gray[100:105, 100:105])
```

#### Short Answers :

##### 1. Pixel Values (Grayscale) :
After converting the image to grayscale, each pixel is represented by a single intensity value instead of three color values. These values range from 0 to 255, where 0 represents black, 255 represents white, and intermediate values represent different shades of gray. The grayscale value is computed as a weighted combination of the original B, G, and R values to reflect human perception of brightness.

##### 2. Information Lost When Removing Color:
Color-related information such as hue and saturation is lost during grayscale conversion. Different colors with similar brightness levels may appear identical in the grayscale image, making it impossible to distinguish them based on color alone.

##### 3. Information Preserved:
Important structural information such as edges, shapes, textures, and shading is preserved because these features mainly depend on intensity changes rather than color. The overall scene and object boundaries remain clearly visible.

##### 4. Why Grayscale Is Used in Early Image Processing:
Grayscale images reduce data complexity by using a single channel instead of three. This makes image processing algorithms faster and more efficient. Many early-stage operations such as edge detection, thresholding, segmentation, and feature extraction rely on intensity variations, making grayscale images a strong foundation for further analysis.



## Task 03 - Resize andd Zoom to study Resolution Effects

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img=cv2.imread("cats.jpg")
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
small= cv2.resize(gray,(100,100))


plt.imshow(small,cmap='gray')
plt.title("Small Grayscale Image")
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img=cv2.imread("cats.jpg")
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
big= cv2.resize(gray,(1000,1000))


plt.imshow(big,cmap='gray')
plt.title("Larger Grayscale Image")
plt.show()
```

#### Short Answers:

##### 1. Resizing the Image:
The grayscale image was first resized down to a smaller resolution (100×100 pixels) and then resized up to a larger size (500×500 pixels). Reducing the image size decreases the number of pixels available to represent visual details. When the image is enlarged again, no new information is created; instead, existing pixel values are stretched to fill the larger area.

##### 2. Visual Details Lost:
Fine details such as sharp edges, small textures, and thin boundaries (for example, mountain edges) disappear when the resolution is reduced. These details cannot be recovered during resizing back to a larger image because the original high-frequency information has already been lost.

##### 3. Pixelation and Blur:
When the low-resolution image is zoomed or enlarged, pixelation becomes visible as block-like square regions, especially along edges. In some cases, blur appears due to interpolation, where pixel values are averaged to create smoother transitions. Both effects reduce edge sharpness and clarity.

##### 4. Effect on Image Processing Tasks:
Reducing resolution negatively affects image processing tasks that rely on fine details, such as edge detection, object recognition, and feature extraction. Important features may become weak or disappear entirely, leading to reduced accuracy. However, lower resolution images require less memory and computational power, which can be beneficial for faster processing in some applications.


## Task 4 - Crop to Focus on Meaningful Regions

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img=cv2.imread("cats.jpg")
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

crop= gray[10:220,120:220]

plt.imshow(crop,cmap='gray')
plt.title("Crop Image")
plt.show()
```

#### Short Answers :

##### 1. Cropping the Image:
In this task, two regions of the grayscale image were cropped: one from the central area and another from the edge region. Cropping selects a specific portion of the image by removing surrounding pixels, allowing focused analysis of a particular region of interest (ROI).

##### 2. Isolating Meaningful Regions:
Cropping helps isolate important features by eliminating irrelevant background information. The central crop usually contains the main subject of the image and shows clearer structural details, while the edge crop may contain partial objects or background elements. This selective focus makes analysis simpler and more efficient.

##### 3. Importance of Spatial Context:
While cropping highlights local details, it can also remove important spatial context. Some features only make sense when viewed in relation to their surroundings. For example, a small texture or edge may be difficult to interpret without knowing its position in the overall scene.

##### 4. Face Crops vs Whole-Scene Understanding:
In face recognition, cropping the face region is useful because it removes unnecessary background and focuses on relevant facial features. However, for scene understanding tasks, the full image is often needed to understand relationships between objects, background, and environment. This shows that the usefulness of cropping depends on the application.

## Task 05 - Flip and Rotate to Explore Viewpoint Changes

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img=cv2.imread("cats.jpg")
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

h,w=gray.shape
M = cv2.getRotationMatrix2D(center=(w//2,h//2), angle=45, scale=1.0) 
rot = cv2.warpAffine(gray, M, (w,h)) 

plt.imshow(rot,cmap='gray')
plt.title("Rotation 45 degree")
plt.show()
```

#### Short Answers :

##### 1. Flipping the Image:
The image was flipped horizontally and vertically. Horizontal flipping preserves most of the visual interpretation because objects remain recognizable and maintain their relative structure. Vertical flipping, however, can distort meaning since it places objects in unnatural orientations, making scenes appear unrealistic and harder to interpret.

##### 2. Rotating the Image:
The image was rotated at different angles such as 45 degrees. Small rotations usually preserve the identity of objects, but larger rotations can reduce clarity and alter the natural viewpoint of the scene. Rotation may also introduce empty regions or cropping at the image boundaries due to geometric transformation.

##### 3. Transformations That Preserve Meaning:
Horizontal flipping and small-angle rotations generally preserve semantic meaning. Objects remain identifiable, and the overall scene can still be understood. These transformations are commonly used because they simulate natural viewpoint changes.

##### 4. Transformations That Distort Meaning:
Vertical flipping and large rotations can distort interpretation by creating unrealistic orientations. Such transformations may confuse both human observers and image processing algorithms if not handled properly.

##### 5. Relation to Data Augmentation:
Viewpoint transformations are widely used in data augmentation to increase the diversity of training data. By applying flips and rotations, models learn to recognize objects under different orientations, improving robustness and generalization.

##### 6. Relation to Real Camera Motion:
Real-world camera motion such as tilting, rotating, or changing viewpoint produces similar transformations in captured images. Simulating these variations helps image processing systems perform reliably under real-world conditions.


## Task 06 - Invert Pixel Intensities (Negative Image)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img=cv2.imread("cats.jpg")
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

inv= cv2.bitwise_not(gray)

plt.imshow(inv,cmap='gray')
plt.title("Negative Image")
plt.show()
```

#### Short Answer:
Pixel intensity inversion disrupts interpretation because it reverses the natural brightness relationships that the human visual system uses to understand scenes. In real-world viewing, bright areas usually indicate light sources or reflective surfaces, while darker areas represent shadows or depth. When an image is inverted, these cues are flipped, causing shadows to appear bright and highlights to appear dark, which makes objects, shapes, and depth harder to recognize and interpret naturally.

However, negative imaging can be beneficial in certain applications where enhancing contrast is more important than realistic appearance. In medical imaging, inversion can help highlight subtle structures or abnormalities that may be difficult to detect in the original image. In forensic analysis, negative images can reveal hidden details such as faint patterns, impressions, or marks that are not easily visible under normal intensity mapping. Thus, while inversion disrupts intuitive understanding, it can be a powerful tool for specialized image analysis tasks.

This spatial arrangement allows the human eye to recognize patterns such as curves and strokes, which together form a meaningful digit shape. Without reshaping, the numbers would remain an unstructured list with no visual meaning.

##### 3. Importance of Correct Reshaping:
Correct reshaping is essential because each pixel value corresponds to a specific location in the image. If the reshaping is incorrect, the spatial structure of the digit is distorted, leading to unrecognizable or misleading images. This directly affects feature extraction and learning performance.

##### 4. Relation to Machine Learning:
Machine learning models depend on correctly structured input data. Incorrect reshaping can confuse the model and significantly reduce classification accuracy.

## Task 08 - Apply a Simple Affine Transformation

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

img=cv2.imread("cats.jpg")
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

pts1 = np.float32([[0,0],[100,0],[0,100]]) 
pts2 = np.float32([[0,0],[80,20],[20,100]]) 
M = cv2.getAffineTransform(pts1, pts2) 
h,w=gray.shape
aff = cv2.warpAffine(gray, M, (w,h)) 

plt.imshow(aff, cmap='gray')
plt.title("Affine Image")  
plt.axis("off")
plt.show()
```

#### Short Answers :
##### 1. Applying Shear/Shift:
In this task, an affine transformation was applied to the grayscale scenic image using shear (or shift). Affine transformations move pixels according to a linear mapping, which can slant, stretch, or shift objects while preserving parallel lines. This operation changes the positions of image points without altering pixel intensities.

##### 2. Effect on Object Shapes:
Shearing or shifting distorts the original shapes of objects in the image. Straight lines may appear slanted, rectangular regions may become parallelograms, and familiar structures like mountains or buildings can appear skewed. Although overall spatial relationships are partially preserved, the geometry of individual objects is altered.

##### 3. Relation to Camera Perspective:
These distortions simulate effects similar to changes in camera perspective, such as tilting the camera or viewing an object from an angle. In real-world photography, objects appear differently depending on the viewpoint, which can stretch or skew shapes. Understanding how shear affects images helps in modeling perspective changes and correcting them in computer vision tasks.
